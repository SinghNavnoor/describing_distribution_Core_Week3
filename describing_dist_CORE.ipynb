```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/data.csv')

df.head()
```

```python
df.columns
```

## Exploring VitD_Levels

```python
V_col = 'VitD_levels'
feature = df[V_col]
V_mean = feature.mean()
V_median = feature.median()
V_std = feature.std()
V_plus_one_std = V_mean + V_std
V_minus_one_std = V_mean - V_std
```

```python
fig,ax = plt.subplots(figsize=(7,6))
sns.histplot(feature, ax = ax, stat='probability');
ax.axvline(V_mean, color = 'red', label = 'Mean');
ax.axvline(V_median, ls = '--', color = 'green', label = 'Median');
ax.axvline(V_plus_one_std, color = 'black', label =f'+1 std = {V_plus_one_std:,.2f}')
ax.axvline(V_minus_one_std, color = 'black', label =f'-1 std = {V_minus_one_std:,.2f}')
ax.axvspan(V_plus_one_std, V_minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{V_col}')
ax.legend();
```

#### This is a continuous value. Vitamin D levels can be described in infinite values.  
#### This feature does not have a Skew. It is very normally distributed. 
#### This feature is also Leptokurtic. We can see that values are concentrated in the center. 

## Exploring Doc_visits

```python
d_col = 'Doc_visits'
d_feature = df[d_col]
d_mean = d_feature.mean()
d_median = d_feature.median()
d_std = d_feature.std()
d_plus_one_std = d_mean + d_std
d_minus_one_std = d_mean - d_std
```

```python

fig,ax = plt.subplots(figsize=(7,6))
sns.histplot(d_feature, ax = ax, stat='probability');
ax.axvline(d_mean, color = 'red', label = 'Mean');
ax.axvline(d_median, ls = '--', color = 'green', label = 'Median');
ax.axvline(d_plus_one_std, color = 'black', label =f'+1 std = {d_plus_one_std:,.2f}')
ax.axvline(d_minus_one_std, color = 'black', label =f'-1 std = {d_minus_one_std:,.2f}')
ax.axvspan(d_plus_one_std, d_minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{d_col}')
ax.legend();
```

#### It is a Discrete feature. The reason being that there are only a set number of times one can visit the doctor in a given time frame. 

#### It is not skewed. It is normally distributed. 

#### This is a Leptokurtic distribution. Looks like that most individuals have have anywhere between 4 - 6 doctors visits.
